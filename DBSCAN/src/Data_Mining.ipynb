#2/ RFM analysis

```python
#Create Total sum columns 
```

```python
df["TotalSum"] = df.Quantity * df.UnitPrice
df.head()
```

```python
#Find Min max of the date 
print('Min_date {} \nMax date {}'. format(min(df.InvoiceDate), max(df.InvoiceDate)))
```

```python
snapshot_date = max(df.InvoiceDate) + dt.timedelta(days = 1)
snapshot_date
```

Group by customer for lastest buying date, number of order, and total amount of money

```python
rfm_data = df.groupby(['CustomerID']).agg({
'InvoiceDate' : lambda x: (snapshot_date - x.max()).days, 
'InvoiceNo'   : 'count',
'TotalSum'   : 'sum'
})
rfm_data.rename(columns = {
                            'InvoiceDate' : 'Recency',
                            'InvoiceNo' : 'Frequency',
                            'TotalSum' : 'MonetaryValue'
                            }, inplace = True)

rfm_data.head()
```

```python
rfm_data
```

```python
rfm_data.to_csv('rfm.csv') # save to CSV
```

#3/ Create dataset for FP growth
- The data will represent the transaction contain each item, since we cluster by customer so we will group by customer with columns is the list of buying item, t indicate is in and ? indicated is not in customer preference. 
- we will create 5 datasets for FP growth 1 is all dataset and 4 represent for each cluster

##3.1/ Full FP growth data

```python
df["Description"] = df["Description"].replace({"[$&+,:;=?@#|'<>.^*()%!-]": ''}, regex=True)
df["Description"] = df["Description"].replace({'"': ''}, regex=True)
df["Description"] = df["Description"].replace({" ": '_'}, regex=True)
```

```python
fp_growth = df.pivot_table(index="CustomerID", columns="Description", aggfunc="count").Quantity
fp_growth
```

```python
fp_growth[~fp_growth.isnull()] = 't'  # not nan
fp_growth[fp_growth.isnull()] = '?'   # nan
```

```python
fp_growth.drop('CustomerID').to_csv("FPgrowth.csv", index=False)#save to CSV. This is our 1st data
```
